```python
# Create a manual DataFrame with MotoGP 2023 statistics
data = {
    "Rider": ["Marc Marquez", "Fabio Quartararo", "Francesco Bagnaia", "Jorge Martin"],
    "Team": ["Repsol Honda", "Yamaha", "Ducati Lenovo", "Pramac Ducati"],
    "Championships": [6, 1, 1, 0],
    "Average_Speed_kmh": [175.2, 178.5, 180.1, 179.8],
    "Active": [True, True, True, True]
}
df1 = pd.DataFrame(data)
df1
```

```python
# write the DataFrame to a CSV file
df1.to_csv("MotoGP_2023_Stats.csv", index=False)
```

```python
# Display specific rows of the DataFrame using slicing
df1[1:3]
```

```python
# Display only the "Team" column from the DataFrame
df1[["Team"]]
```

```python
# max, min, mean, std of Average_Speed_kmh
print(df1.Average_Speed_kmh.max(), df1.Average_Speed_kmh.min(), df1.Average_Speed_kmh.mean(), df1.Average_Speed_kmh.std())
```

```python
# Display the DataFrame's descriptive statistics
df1.describe()
```

```python
# Filter the DataFrame to show only active riders
df1[df1.Active == True]
```

```python
# filter the DataFrame to show riders with the maximum average speed
df1[df1.Average_Speed_kmh == df1.Average_Speed_kmh.max()]
```
