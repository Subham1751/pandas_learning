```python
import pandas as pd
```

```python
df = pd.read_csv("weather.csv")
df
```

```python
# Pivot the DataFrame to have 'Date' as index and 'City' as columns with 'Wind_Speed' as values
df.pivot(index = "Date", columns = "City" , values = "Wind_Speed")
```

```python
df = pd.read_csv("weather2.csv")
df
```

```python
# use pivot_table to find the average temperature for each city on each date
df.pivot_table(index = "City" , columns = "Date" , aggfunc="mean")
```

```python
df = pd.read_csv("weather2.csv")
df["Date"] = pd.to_datetime(df["Date"])
df
```

```python
# use grouper to group by datetime and frequency 
df.pivot_table(index = pd.Grouper(freq = "M" , key = "Date"), columns = "City")
```
