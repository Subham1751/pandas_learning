```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# sample data for India
data_india = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Bangalore", "Kolkata"],
    "Temperature": [30, 28, 25, 27],
    "Humidity": [60, 70, 65, 75]
})
data_india
```

```python
# sample data for Canada
data_canada = pd.DataFrame({
    "City": ["Toronto", "Vancouver", "Montreal", "Calgary"],
    "Temperature": [-5, 10, 0, 5],
    "Humidity": [80, 75, 85, 70]
})
data_canada
```

```python
# Concatenating the two DataFrames
df = pd.concat([data_india, data_canada], ignore_index=True)
df
```

```python
# Concatenating with keys to distinguish between the two DataFrames
df1 = pd.concat([data_india, data_canada], keys=['India', 'Canada'])
df1
```

```python
# Concatenating along columns
df = pd.concat([data_india, data_canada], axis=1)
df
```

```python
weather_data = pd.Series(["Sunny", "Rainy", "Cloudy", "Windy"], name="Weather")
weather_data
```

```python
# joining dataframes with series function
df = pd.concat([data_india,weather_data], axis=1)
df
```
