```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Bangalore", "Chennai", "Hyderabad"],
    "Temperature": [30, 32, 28, 33, 29],
})
df1
```

```python
df2 = pd.DataFrame({
    "City": ["Mumbai", "Bangalore", "Delhi" , "Shimla"],
    "Humidity": [70, 65, 80, 60],
})
df2
```

```python
# Merging the two DataFrames on the "City" column using an inner join
df = pd.merge(df1, df2, on="City" , how="inner")
df
```

```python
# Merging the two DataFrames on the "City" column using an outer join and adding an indicator column
df = pd.merge(df1, df2, on="City" , how="outer" , indicator=True)
df
```

```python
# Merging the two DataFrames on the "City" column using a left join
df = pd.merge(df1, df2, on="City" , how="left")
df
```

```python
# Merging the two DataFrames on the "City" column using a right join
df = pd.merge(df1, df2, on="City" , how="right")
df
```
